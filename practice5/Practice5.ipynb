### Fetching Dataset

````python
from ucimlrepo import fetch_ucirepo 
````

````python
# fetch dataset 
statlog_german_credit_data = fetch_ucirepo(id=144) 
````

````python
# data (as pandas dataframes) 
X = statlog_german_credit_data.data.features 
y = statlog_german_credit_data.data.targets 
````

````python
# metadata 
print(statlog_german_credit_data.metadata) 
````

````python
# variable information 
print(statlog_german_credit_data.variables) 
````

### Training model

````python
y = y.replace({1: 1, 2: 0})  # 1 = good, 0 = bad
````

````python
import pandas as pd

df = pd.concat([X, y], axis=1)
df.rename(columns={'class': 'target'}, inplace=True)
````

Dataset encodes sex like:

    A91, A93, A94 → male
    A92, A95 → female
So:

````python
def map_sex(x):
    if x in ['A91', 'A93', 'A94']:
        return 'male'
    else:
        return 'female'

df['Sex'] = df['Attribute9'].apply(map_sex)
````

````python
df['Age_group'] = df['Attribute13'].apply(lambda x: '<30' if x < 30 else '>=30')
````

````python
df_encoded = pd.get_dummies(df, drop_first=True)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X = df_encoded.drop(columns=['target'])
y = df_encoded['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LogisticRegression(max_iter=5000)
model.fit(X_train, y_train)
````

````python
X_original = df.copy()
````

````python
X_train, X_test, y_train, y_test, X_train_orig, X_test_orig = train_test_split(
    df_encoded.drop(columns=['target']),
    df_encoded['target'],
    X_original,
    test_size=0.2
)
````

### Fairness evalution

````python
from sklearn.metrics import accuracy_score, recall_score

def evaluate_group(feature):
    groups = X_test_orig[feature].unique()
    
    for g in groups:
        idx = X_test_orig[feature] == g
        
        X_g = X_test[idx]
        y_g = y_test[idx]
        
        y_pred = model.predict(X_g)
        
        acc = accuracy_score(y_g, y_pred)
        recall = recall_score(y_g, y_pred)
        
        print(f"{feature} = {g}")
        print(f"Accuracy: {acc:.3f}, Recall: {recall:.3f}")
        print("------")
````

#### Sensitive Features

````python
evaluate_group('Sex')
evaluate_group('Age_group')
````

## Answer key questions

**Q1: Does model performance differ?**
```
There are significant differences across both Proxy and Sensitive feature groups.
Performance varies notably in both accuracy and recall.
```

**Q2: Which groups are worse?**  
The most disadvantaged groups are:
- Age < 30 (younger individuals)
- Female applicants (moderate effect)
- Housing A153 (free housing)
- Job A171 (unemployed), which shows extremely poor performance

**Q3: Do proxy features reveal additional bias?**  
Yes, proxy features reveal significantly stronger bias.  
Groups based on Job and Housing show much larger performance gaps
compared to Sex and Age, indicating indirect discrimination.  

**Q4:Why might these differences occur?**

These differences occur because the model learns patterns
from historical data, where socio-economic factors such as
employment and housing status are correlated with credit risk.
As a result, the model captures these correlations, even if
they lead to unfair outcomes.

**Q5: Why is this critical in real-world systems?**  
In real-world systems like credit scoring, such biases can lead
to unfair treatment of certain groups. For example, younger or
unemployed individuals may be denied loans more frequently,
even when they are actually reliable borrowers. This can result
in financial exclusion and reinforce existing inequalities.
