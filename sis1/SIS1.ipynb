### 1.2 A2. Descriptive Statistics (0.5)

#### 1 — Select Numeric Columns

```python
# Select numeric columns only
numeric_df = df.select_dtypes(include=[np.number])

numeric_df.columns
```

#### 2 — Compute Descriptive Statistics

```python
# Basic statistics
mean = numeric_df.mean()
median = numeric_df.median()
mode = numeric_df.mode().iloc[0]
data_range = numeric_df.max() - numeric_df.min()
variance = numeric_df.var()
std_dev = numeric_df.std()
iqr = numeric_df.quantile(0.75) - numeric_df.quantile(0.25)

stats_df = pd.DataFrame({
    "Mean": mean,
    "Median": median,
    "Mode": mode,
    "Range": data_range,
    "Variance": variance,
    "Std Dev": std_dev,
    "IQR": iqr
})

stats_df
```

#### 3 — Create a boxplot; mark suspected outliers and list their values

##### 1. Boxplot with Outliers

```python
# 1. Clean the price column (Remove $ and ,)
# We use .str.replace to strip the non-numeric characters
df['price'] = df['price'].replace(r'[\$,]', '', regex=True)

# 2. Convert to numeric (errors='coerce' turns problematic rows into NaN)
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# 3. Drop rows where price is now missing so the plot doesn't break
df = df.dropna(subset=['price'])

# 4. Now plot again
plt.figure(figsize=(10, 5))
sns.boxplot(x=df["price"])
plt.title("Cleaned Boxplot of Airbnb Prices")
plt.xlabel("Price ($)")
plt.show()
# plt.figure(figsize=(8,5))
# sns.boxplot(x=df["price"])
# plt.title("Boxplot of Airbnb Prices")
# plt.xlabel("Price")
# plt.show()
```

##### 2. Identify Outliers (IQR Method)

```python
print(df["price"].describe())

Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)
IQR_price = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR_price
upper_bound = Q3 + 1.5 * IQR_price

outliers = df[(df["price"] < lower_bound) | (df["price"] > upper_bound)]
# print(lower_bound)
# print(upper_bound)
outliers["price"]
```
